# add clustered result in csv file
after estimating each jaccard, edit, cosine similarity clustering, add cluster number of each email into our csv file

```python
import json
import os
import pandas as pd
```

```python
cwd = os.getcwd()
```

```python
fileCos = cwd + "/cosine_circle.json" ## name the file as cosine_circle before execute it.
fileEdi = cwd + "/edit_circle.json" ## name the file as cosine_circle before execute it.
fileJac = cwd + "/jaccard_circle.json" ## name the file as cosine_circle before execute it.
```

```python
featuretable = cwd + "/featureTable.csv"
featureTable = pd.read_csv(featuretable)
```

```python
with open(fileCos, encoding='utf-8') as openFile:
    cosineCircle = json.load(openFile)

with open(fileEdi, encoding='utf-8') as openFile:
    editCircle = json.load(openFile)

with open(fileJac, encoding='utf-8') as openFile:
    jaccardCircle = json.load(openFile)
```

```python

lenCos = len(cosineCircle["children"])
lenEdi = len(editCircle["children"])
lenJac = len(jaccardCircle["children"])

rows_Cos = []
rows_Edi = []
rows_Jac = []


for i in range(lenCos):
    totalNoEmail = len(cosineCircle["children"][i]["children"])
    for j in range(totalNoEmail):
        childrenName = cosineCircle["children"][i]["children"][j]["name"]
        if childrenName.__contains__(".json"):
            file = childrenName[childrenName.find("/")+1:childrenName.find('.json')] + ".json"
        name = cosineCircle["children"][i]["name"]
        row = {"file":file,"name":name}
        rows_Cos.append(row)
clusterTableCos = pd.DataFrame.from_dict(rows_Cos, orient='columns')


for i in range(lenEdi):
    totalNoEmail = len(editCircle["children"][i]["children"])
    for j in range(totalNoEmail):
        childrenName = editCircle["children"][i]["children"][j]["name"]
        if childrenName.__contains__(".json"):
            file = childrenName[childrenName.find("/")+1:childrenName.find('.json')] + ".json"
        name = editCircle["children"][i]["name"]
        row = {"file":file,"name":name}
        rows_Edi.append(row)
clusterTableEdi = pd.DataFrame.from_dict(rows_Edi, orient='columns')


for i in range(lenJac):
    totalNoEmail = len(jaccardCircle["children"][i]["children"])
    for j in range(totalNoEmail):
        childrenName = jaccardCircle["children"][i]["children"][j]["name"]
        if childrenName.__contains__(".json"):
            file = childrenName[childrenName.find("/")+1:childrenName.find('.json')] + ".json"
        name = jaccardCircle["children"][i]["name"]
        row = {"file":file,"name":name}
        rows_Jac.append(row)
clusterTableJac = pd.DataFrame.from_dict(rows_Jac, orient='columns')
```

```python
for i in range(len(featureTable)):
    file = str(i) + ".json"
    clusterCos = clusterTableCos.loc[clusterTableCos["file"]==file,"name"].values[0]
    clusterEdi = clusterTableEdi.loc[clusterTableEdi["file"]==file,"name"].values[0]
    clusterJac = clusterTableJac.loc[clusterTableJac["file"]==file,"name"].values[0]
    featureTable.loc[featureTable["filename"]==file,"cosineCluster"] = clusterCos
    featureTable.loc[featureTable["filename"]==file,"ediCluster"] = clusterEdi
    featureTable.loc[featureTable["filename"]==file,"jaccardCluster"] = clusterJac
```
